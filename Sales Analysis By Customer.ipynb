```python
## Importing use full python libraries for analysis

import numpy as np
import pandas as pd 
import seaborn as sna
import matplotlib.pyplot as plt
```

```python
#Importing the data that will be analyzed
sales_df = pd.read_excel('Mock Actual Sales.xlsx')
```

```python
#Showing the heading of the data frame to see what we are working with

sales_df.head()
```

# Adding columns

```python
## Adding a Columns for Gross_Price/Unit, Net_Price/Unit, and COGS/Unit

sales_df['Gross Price/Unit'] = sales_df['Gross Sales']/sales_df['Quantity']
sales_df['Net Price/Unit'] = sales_df['Net Sales']/sales_df['Quantity']
sales_df['COGS/Unit'] = sales_df['COGS']/sales_df['Quantity']

sales_df.head()
```

# Grouping Data

```python
sales_df.groupby('Customer').sum()
```

# Analysis through Visualization

```python
y = sales_df.groupby('Customer').sum()

x = range(4)
x_label = ['Customer A', 'Customer B', 'Customer C', 'Customer D']

plt.bar(x, y['Quantity'])
plt.xticks(x, x_label, rotation='vertical')
plt.title('Quantity by Customer')

plt.show()

plt.bar(x, y['Gross Sales'])
plt.xticks(x, x_label, rotation='vertical')
plt.title('Gross Sales by Customer')

plt.show()

plt.bar(x, y['Net Sales'])
plt.xticks(x, x_label, rotation='vertical')
plt.title('Net Sales by Customer')

plt.show()

plt.bar(x, y['COGS'])
plt.xticks(x, x_label, rotation='vertical')
plt.title('COGS by Customer')

plt.show()
```
